```python
x1 = Node(1.0, label="x1")
x2 = Node(-2.0, label="x2")

w1 = Node(3.0, label="w1")
w2 = Node(-1.5, label="w2")

b1 = Node(0.1, label="b1")

x1w1 = x1 * w1
x1w1._label = "x1w1"

x2w2 = x2 * w2
x2w2._label = "x2w2"

xw = x1w1 + x2w2
xw._label = "xw"

xwb = xw + b1
xwb._label = "xwb"

y = xwb.relu()
y._label = "y"
```

```python
y.backward()
```

```python
Visualize.draw(y, rankdir="TB")
```
